```python
import sys, os
import cv2
import numpy as np

def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict


def main():
    train_ds_file = os.path.join(clean_path, 'train.txt')
    test_ds_file = os.path.join(clean_path, 'val.txt')
    tree_file = os.path.join(clean_path, 'tree.txt')
    
    if os.path.isfile(train_ds_file):
        os.remove(train_ds_file)
        
    if os.path.isfile(test_ds_file):
        os.remove(test_ds_file)
    
    # process training data
    blob = unpickle(cifar_train_ds_path)
    print(blob.keys())
    n_data = len(blob[b'data'])
    tree_nodes = {}
    
    total_data = 0
    with open(train_ds_file, 'w') as f:
        for i in range(n_data):
            total_data+=1

    for i in range(n_data):
        data = blob[b'data'][i]
        im_data = np.zeros((32,32,3), dtype=np.uint8)
        im_data[:,:,2] = data[:1024].reshape(32,32)
        im_data[:,:,1] = data[1024:2048].reshape(32,32)
        im_data[:,:,0] = data[2048:3072].reshape(32,32)
        sup_label = blob[b'coarse_labels'][i]
        child_label = blob[b'fine_labels'][i]
        if sup_label not in tree_nodes:
            tree_nodes[sup_label] = []
        if child_label not in tree_nodes[sup_label]:
            tree_nodes[sup_label].append(child_label)
        im_path = os.path.join(train_path, '%d.png'%i)
        rel_im_path = 'train/%d.png'%i
        cv2.imwrite(im_path, im_data)
        if os.path.isfile(train_ds_file):
            with open(train_ds_file, 'a+') as f:
                f.write('%s, %d\n'%(rel_im_path, child_label))
        else:
            with open(train_ds_file, 'w') as f:
                f.write('%s, %d\n'%(rel_im_path, child_label))
        if i%1000 == 0:
            print('processed %d/%d'%(i, total_data))
            
        
    # process testing data
    blob = unpickle(cifar_test_ds_path)
    n_data = len(blob[b'data'])
    
    total_data = 0
    with open(test_ds_file, 'w') as f:
        for i in range(n_data):
            total_data+=1
    
    for i in range(n_data):
        data = blob[b'data'][i]
        im_data = np.zeros((32,32,3), dtype=np.uint8)
        im_data[:,:,2] = data[:1024].reshape(32,32)
        im_data[:,:,1] = data[1024:2048].reshape(32,32)
        im_data[:,:,0] = data[2048:3072].reshape(32,32)
        sup_label = blob[b'coarse_labels'][i]
        child_label = blob[b'fine_labels'][i]
        im_path = os.path.join(train_path, '%d.png'%i)
        rel_im_path = 'test/%d.png'%i
        cv2.imwrite(im_path, im_data)
        if os.path.isfile(test_ds_file):
            with open(test_ds_file, 'a+') as f:
                f.write('%s, %d\n'%(rel_im_path, child_label))
        else:
            with open(test_ds_file, 'w') as f:
                f.write('%s, %d\n'%(rel_im_path, child_label))
        if i%1000 == 0:
            print('processed %d/%d'%(i, total_data))
            
            
    with open(tree_file, 'w') as f:
        for sup_cls in tree_nodes.keys():
            for child_cls in tree_nodes[sup_cls]:
                f.write('%d, %d\n'%(sup_cls, child_cls))
                

if __name__ == '__main__':
    cifar_ds_path = '/datasets/vision/cifar-100-python'
    cifar_train_ds_path = os.path.join(cifar_ds_path, 'train')
    cifar_test_ds_path = os.path.join(cifar_ds_path, 'test')
    clean_path = '/datasets/vision/cifar100_clean'
    
    if not os.path.isdir(clean_path):
        os.mkdir(clean_path)
    
    train_path = os.path.join(clean_path, 'train')
    test_path = os.path.join(clean_path, 'test')
    
    if not os.path.isdir(train_path):
        os.mkdir(train_path)
    if not os.path.isdir(test_path):
        os.mkdir(test_path)
    
    main()
    print('Done')
```
